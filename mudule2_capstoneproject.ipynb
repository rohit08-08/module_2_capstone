## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


 ## Step 1: Import required libraries
import pandas as pd

# Step 2: Load the datasets
titles = pd.read_csv('/content/drive/MyDrive/titles.csv')
credits = pd.read_csv('/content/drive/MyDrive/credits.csv')

# Step 3: Merge both datasets using 'id'
df = pd.merge(titles, credits, on='id', how='left')

# Step 4: Drop duplicate rows
df.drop_duplicates(inplace=True)

# Step 5: Fill missing values
df['age_certification'].fillna('Unknown', inplace=True)
df.dropna(subset=['imdb_score', 'tmdb_score'], inplace=True)

# Step 6: Convert genre and country columns to lists
import ast

df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df['production_countries'] = df['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# Step 7: Add useful new columns
df['is_movie'] = df['type'].apply(lambda x: 1 if x.lower() == 'movie' else 0)

df['release_decade'] = (df['release_year'] // 10) * 10

# Step 8: Clean data types and remove bad rows
df = df[pd.to_numeric(df['imdb_score'], errors='coerce').notnull()]
df = df[pd.to_numeric(df['runtime'], errors='coerce').notnull()]
df['runtime'] = df['runtime'].astype(float)

# Step 9: Final check
print(" Final cleaned dataset shape:", df.shape)


````

### What all manipulations have you done and insights you found?

# **Insights Found**

**Missing values** were mostly in age_certification. Handling them ensured we can still include those rows in analysis without deleting useful data.

**IMDb & TMDb scores** cleaning removed invalid entries (like NaN or wrong types). This avoids misleading averages or false trends.

**Genres & countries** as lists let us study multi-genre shows and cross-country productions. For example, a movie can be both Action and Comedy.

**New features** (is_movie, release_decade) revealed patterns:

**TV shows vs. Movies** can be compared directly.

**Popular decades of releases** can be analyzed (e.g., more content released in 2010s).

**Runtime cleaning** allowed us to study average movie length, detect outliers, and compare runtimes by genre/type.
